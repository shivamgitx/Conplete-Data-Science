### map() function

The Python map() function applies a given function to each item in an iterable (like a list, tuple, or string) and returns a map object (an iterator) containing the results. It offers a concise way to perform data transformations without explicit for loops.

#### syntax:

map(function_name,iterable )

```python
def square(num):
    return num ** 2

square(5)    
```

```python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))          #syntax= map(function,iterable)
                                   #iterable=collection of items
```

```python
##lambda function with map

numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
```

```python
## map multiple iterables 

numbers1=[1,2,3]
numbers2=[4,5,6]

add_numbers=map(lambda x,y:x+y,numbers1,numbers2)

print(list(add_numbers))
```

```python
## map() to convert a list of strings to integers
## use mapto convert strings to integers

str_numbers=['1','2','3','4','5']
int_numbers=list(map(int,str_numbers))

print(int_numbers)
```

```python
words=['apple','banana','cherry']
upper_words=list(map(str.upper,words))

print(upper_words)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'Shivam','age':24},
    {'name':'Yush','age':26},
]

list(map(get_name,people))
```
