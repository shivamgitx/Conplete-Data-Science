## Lmbda Functions in Python
A lambda function in Python is a small, anonymous function(function without name) that can take any number of arguments but is restricted to having only one expression. It is defined using the lambda keyword and is commonly used for short, single-use operations, often as an argument to higher-order functions like map(), filter(), and sorted()


-anonymous function(function without name)


-only one expression (only one logic)


-using lanmbda keyword

```python
# Syntax
lambda arguments: expression
```

```python
def addition(a,b):#function
    return a + b

addition(2,3)
```

```python
addition=lambda a,b:a+b #lambda function or anonymous function 
type(addition)
print(addition(1,2))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)    
```

```python
even1=lambda num:num%2==0
even1(24)
```

```python
def addition(a,b,c):
    return(a+b+c)

addition(1,2,3)
```

```python
addition=lambda a,b,c:a+b+c
addition(1,2,3)
```

```python
## map()--applies a function to all items in list

numbers=[1,2,3,4,5]

def square(num):       #example
    return num**2

square(2)
```

```python
list(map(lambda x:x**2,numbers))
```
