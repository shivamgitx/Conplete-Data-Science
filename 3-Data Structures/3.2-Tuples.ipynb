# Tuples

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutabilitymakes them different.

```python
## creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"Hello World", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple element
numbers
```

```python
print(numbers[2])
print(numbers[-1])
print(numbers[1:4])
print(numbers[1:])
print(numbers[:4])
print(numbers[:])
```

```python
## Tuples Operations

concatenation_tuples=numbers+mixed_tuple
concatenation_tuples
```

```python
mixed_tuple*3
```

```python
## Immutable nature of tuples
## Tuples are immutable, meaning their elements can not be change once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Shivam"
print(lst)
```

```python
## Tuple Methods

print(numbers.count(1))
print(numbers.index(1))
```

```python
## Pcking and unpacking tuples

packed_tuple=1,"Hello", 3.14
print(packed_tuple)



```

```python
## unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## unpacking with *

numbers=(1,2,3,4,5,6)
first,*middle,last=numbers

print(first)
print(middle)
print(last)
```

```python
## Nexted tuple
## nested list

lst=[[1,2,3,4],[5,6,7,8],[1,"hello",3,14,"c"]]
lst[0][0:3]
```

```python
nested_tuple = ((1,2,3,), ("a","b","c"),(True,False))

## access the element s inside a tuple


print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
