## Dictionaries

Dictionaries are unorderd collections of items. They store data in key-value pairs. Keys ust be unique and immutable(e.g., strings, numbers, or tuples), while values can be of any type.

```python
##creating Dictionary
empty_dict={}
print(type(empty_dict))
```

```python
student={"name":"Shivam", "age":23, "grade":24}
print(student)
print(type(student))
```

```python
# single key is always used
student={"name":"Shivam", "age":23, "name":24}
print(student)
```

```python
## accessing dictionary elements
student={"name":"Shivam", "age":23, "grade":24}
print(student)

print(student['grade'])
print(student['age'])

## accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## modifying dictionary element
## dictionaries are mutable , so you can add, update, delete elements

print(student)
```

```python
student["age"]=33  ##update value for a key
print(student)
student["address"]="India"  ##added a new key and value
```

```python
del student['grade']  ##delete key and value pair
print(student)
```

```python
## Dictionary method

keys=student.keys()  ## get all keys
print(keys)

values=student.values()  ## get all values
print(values)

items=student.items()  ## get all key value pairs
print(items)
```

```python
## Shallow Copy
student_copy=student
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()  ##shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="shivam2"
print(student_copy1)
print(student)
```

```python
## Iterating over Dictionaries
## You can use loops to iterate over dictionaries, keys, values, or items

## Iterarting over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for values in student.values():
    print(values)
```

```python
## Iterate over key value pairs
for key,values in student.items():
    print(f"{key}:{values}")
```

```python
## nested dictionaries
students={
    "student1":{"name":"Shivam", "age":24},
    "student2":{"name":"Yush", "age":26}
}
print(students)

```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension
squares={x:x**2 for x in range(5)}
squares
```

```python
## condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## merge two dictionaries into one
dict1={"a":1,"b":2}
dict2={"c":3,"d":4}
dict3={**dict1,**dict2}
print(dict3)
```
