## Sets

Sets are a built-in data type in Python used to store collections of unique items.They are unordered, meaning that the elements o not follow a specific order, and do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
## create a set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,6,5,4,5,6])
print(my_empty_set)
```

```python
## Basic Sets Operation
## Adding and Removing elements

my_set.add(7)
print(my_set)   
my_set.add(7)
print(my_set)  
```

```python
## Remove element from a set

my_set.remove(3)
my_set
```

```python
my_set.discard(11)  ##if their is an element it will remove otherwise it will ot give any error
my_set
```

```python
## pop method (removing random element)

removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
## clear all the elements
my_set.clear()
print(my_set)
```

```python
## set membership test(is tyhe element in the setor or not)

my_set={1,2,3,4,5}
print(3 in my_set )
print(11 in my_set )
```

```python
## mathematical operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

union_set=set1.union(set2)
print(union_set)
 
```

```python
## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)



set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

#difference
print(set1.difference(set2))
```

```python
set1
```

```python
set2.difference(set1)
```

```python
## symmetric difference
set1.symmetric_difference(set2)
```

```python
## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}
print(set1.issubset(set2))
print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set2.issuperset(set1))
```

```python
lst=[1,2,2,3,4,4,5]

set(lst)
```

```python
## Counting unique words in text

text="In this tutorial we are discussed about sets"
words=text.split()

## convert listof words to set to get unique word

unique_word=set(words)
print(unique_word)
print(len(unique_word))
```
