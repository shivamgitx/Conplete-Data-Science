# Syntax and Semantics in Python


Syntax is the set of rules for structuring sentences or code (grammer/order), while semantics refers to the actual meaning or logic behind those words or symbols.
Syntax:How
Semantics:What

```python
## Basic Syntax Rules
## 1. Case Sensitivity - Python is case-sensitive

name="Shivam"
Name="Prajapati"

print(name)  # Output: Shivam
print(Name)  # Output: Prajapati

```

```python
## 2. Indentation - Python uses indentation to define code blocks. Consistent use of spaces or tabs is crucial.
## Indentation is most important in python Codings

age=24
if age>=18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

```python
## Single line comment
print("Hello World!")
```

```python
## Line continuation

total = 1 + 2 + 3 + 4 + \
        5 + 6 + 7 + 8 

print("Total:", total)
```

```python
## Multiple Statements in a Single Line

x=5; y=20;z=x+y
print(z)
```

```python
## Understand Semantics in Python
# variable assignment

age=24 ## age is an integer
name="Shivam" ## name is a string
```
