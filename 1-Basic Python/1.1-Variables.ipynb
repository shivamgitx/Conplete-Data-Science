# Variables

In Python, variables are essentially symbolic names that serve as references or pointers to objects stored in memory. Unlike many other languages, you do not need to declare a variable's type; it is created the moment you assign a value to it using the = operator. 


```python
## Declaring and Assigning Variables

age=24
height=5.9
name="Shivam"
is_student=True

## printing variables

print("age :",age)
print("height :",height)
print("name :",name)
print("is_student :",is_student)
```

```python
## Naming Conventions
## VAriables bames should be descriptive
## They must start with a letter or an underscore (_) and contains letters, digits, or underscores.
## Variable names are case-sensitive (age and Age are different variables).
## Valid variable names: my_variable, _data, age2
## Invalid variable names: 2age, my-variable, my variable

First_name = "Shivam" 
Last_name = "Prajapati"
```

```python
## case sensitivity

name="Shivam"
Name="Prajapati"
```

```python
## understanding Varaibles types
## Python is dynamically typed language: type of variable is determined at run time based on the assigned value.

age=24 #int
height=5.9 #float
name="Shivam" #str
is_student=True #bool

print(type(age))  
print(type(height))
```
