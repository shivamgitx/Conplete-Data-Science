```python
## while loop

## The while loop continues to execute as long as the condition is TRUE.

count=0

while count<5:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## Break
## The break statement exit the loop permaturely

## Break the statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## countinue 

## The continue statement skips he current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## Pass

## The Pass statement is a null operation ; it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
for i in range(5):
    if i==3:
        print("the number is",i)
        pass
    print(i)
```

```python
## Nested Loops

## a loop inside a loop 

for i in range(3):
    for j in range(2):
        print(i,j)
```

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i}and j:{j}") ## print(f"i:{i}and j:{j}") (i,j)
```

```python
## Examples - Calculate the sum of first N natural numbers using a whie nd for loop

## while loop

n=10
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1

print("Sum of first 10 natural numbers",sum)
```
