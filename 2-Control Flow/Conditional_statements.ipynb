```python
## else
## The else statement executes a block of code if the condition in the f stement is false.

age=16

if age>=18:
    print("you can vote")
else:
    print("can't vote")
```

```python
## elif
## the elif statement allows you to check multiple conditions . It stands for "else if "

age=20

if age<13:
    print("you are child")
elif age<18:
    print("you are teenager")
else:
    print("you are adult")              
```

```python
## Nested Conditional Statements

##you can place one or more if, elif, or else statements nside another if, elif, else statement to create a nested conditional statement.

## number even, odd , negative 

num=int (input("Enter he number"))

if num>=0:
    print("The number s positive")
    if num%2==0:
        print("THe number is even")
    else:
        print("The number is odd")

else:
    print("The number is zero or negative")        

```

```python
## Practical Examples

## Determine f a year is a leap year using nested condition statement

year=int(input("Enter the year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print("The year is a leap year")
        else:
            print("The year is not a leap year")
    else:
        print(year,"is a leap year")

else:
    print(year,"is not a leap year")                
```

```python
## Assignment
## Simple Calculator Program 
## Take user input

num1 = float(input("Enter the first number"))
num2 = float(input("Enter the second number"))
operation = input("Enter the (+, _, *, /):")

## Perform the requested operation

if operation == "+":
    print(num1 + num2)
elif operation == "-":
    print(num1 - num2)
elif operation == "*":
    print(num1 * num2)
elif operation == "/":
    if num2!=0:
        print(num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operation")
```
