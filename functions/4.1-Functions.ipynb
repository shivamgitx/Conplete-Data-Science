## Functions

 Function is a block of code that performs a specific task. Functions help in organized code, reusing code, improving readability.

```python
## syntax
def function_name (parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
## why functions?
num=24
if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd number"""
    if num%2==0:
        print("Even")
    else:
        print("Odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## functions with multiple parameters

def add(a,b):
    return a+b

result=add(2,3)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} wlecome to the page")

greet("Shivam")    
```

```python
## variable length argument
## positional and keyword arguments

def print_numbers(*args):     # here youb can also write any name like *shivam. *args will having multiple parameters
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,"shivam")  # *args is basically refering to all positional arguments
```

```python
## Positional arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,"shivam") 
```

```python
### keywords arguments

def print_details(**kwargs):  # keywords have ** and in it is in form of key value pairs
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="shivam",age=32,country="India")
```

```python
## using both positional and keyword arguments

def print_details(*args,**kwargs):

    for val in args:
        print(f"Positional arguments: {val}")


    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"shivam",name="shivam",age=32,country="India")
```

```python
## Return statements:
# it is used to make our function return something after execution of entire function

def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
## Return multiple parameters

def multiply(a,b):
    return a*b,b

multiply(2,3)
```
