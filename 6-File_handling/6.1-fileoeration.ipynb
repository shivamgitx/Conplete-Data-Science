### File Operation - Read and Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, oth text and binary. This lesson will cover the basics of the file handling, including reading and writing text files and binary files.

```python
### Read a whole file

with open('example.txt','r') as file:
    content =file.read()
    print(content)
```

```python
## read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
## read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  #strip() remove the white space or you can say removes newline character.
```

```python
## Writing a file(overwriting)
with open('example.txt','w')as file:
    file.write('I want to become a data scientist')
    print(file)
```

```python
##write a file withoput overwriting(append)
with open('example.txt','a')as file:
    file.write('appendoing new line\n')
```

```python
## writung a list of lines to a file
lines=['first line\n','second line\n','third line\n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
### Binary files
# basucally a bin file is a file that contains binary data such as images, audio, video, etc.

# writing to a binary file
data= b'\x00\x01\x02\x03\x04\x05'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
# reading to a binary file
with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
### Read the content from the source text file and write to the destination text file

#copying text file
with open('example.txt','r')as source_file:
    content=source_file.read()

with open('destination.txt','w')as destination_file:
    destination_file.write(content)    
```

```python
### Read a text file and count the number of line, words and characters.

def count_text_file(file_path):
    with open(file_path,'r')as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split())for line in lines)
        char_count=sum(len(line)for line in lines)
    return line_count, word_count, char_count

file_path='example.txt'
lines, words, characters=count_text_file(file_path)
print(f'lines: {lines},Words: {words},Characters: {characters}')
```

#### The w+ mode in python is used to open a file for both reading and writing. If he file is does not exist, it will be created. If the file exists, its content is truncated (i.e.,the file is overwritten)
