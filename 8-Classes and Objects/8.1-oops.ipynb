### Classes and Objects

Object-Oriented Programming (oop) is a programming paradigm that uses "objects" to design applications and computer programs.OOP allows for modelling real-world scenarios using classes and objects. This lesson covers the basic of creating classes and objects, including instanmce variables and methods.

```python
### A class is a blueprint for creating objects.

class Car:    #here clar is a class or a blueprint
    pass

audi=Car()
bmw=Car()

print(type(audi))
```

```python
### InstAnce variable and methods
class Dog:
    ##constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

## create objects
dog1=Dog("tom",3)
print(dog1)
print(dog1.name)
print(dog1.age) 
```

```python
dog2=Dog("lucy",4)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
## Define a class with instance methods

class Dog:
    def __init__(self,name,age):      #init is a constructor ,It assigns values to the object’s variable 
        self.name=name
        self.age=age

    def bark(self):                             #instance method
        print(f"{self.name} says woof woof")    

dog1=Dog("tom",3)
dog1.bark()   
     
```

class Dog → blueprint for creating dogs

__init__ → constructor (runs automatically) 

self → current object

name, age → inputs (parameters)

self.name, self.age → object properties

```python
dog2=Dog("lucy",4)
dog2.bark()
```

```python
class Car:
    def __init__(self, model, year):
        self.model = model
        self.year = year

    def start(self):
        print(f"Starting the {self.model} from {self.year}")

car1 = Car("BMW", 2022)
car1.start()
```

```python
## modelling a bank account

## define a. lass for bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount } is deposited .New balance is {self.balance}")

    def withdraw(self,amount):    
        if amount>self.balance:
            print("insufficient balance")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn .New balance is {self.balance}")

    def get_balance(self):
        return self.balance

##  create an object/account

account=BankAccount("shinchan",5000)
print(account.balance)            
```

```python
## call instance methods
account.deposit(100)
```

```python
account.withdraw(300)
```

```python
account.get_balance()
```
