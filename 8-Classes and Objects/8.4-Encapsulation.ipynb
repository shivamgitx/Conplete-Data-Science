## Encapsulation and Abstraction

Encapsulation and abstraction are two of the "four pillars" of Object-Oriented Programming (OOP) in Python. While they are often used together to build robust software, they solve different problems: encapsulation is about "hiding data" to protect it, whereas abstraction is about "hiding complexity" to make a system easier to use. 

## Encapsulation

```python
### encapsulation with getter and setter methods
## Public, protected and private variables or access modifiers

class Person:
    def __init__(self, name, age):
        self.name = name # public variable
        self.age = age # public variable

def get_name(person):
    return person.name

person=Person("John", 30)       
get_name(person)
```

```python
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name # private variable
        self.__age = age # private variable
        self.gender = gender # public variable


def get_name(person):
    return person.__name

person=Person("John", 30, "male")    
get_name(person)  ##you cant access bcz its private variable
```

```python
person=Person("John", 30, "male")
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name # protected variable
        self._age = age # protected variable: we cannt access it outside the class but can access it with derived class
        self.gender = gender # public variable


#derived class
class Employee(Person):        #inheritance it inherit person
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)
    

employee=Employee("John", 30, "male")
print(employee._name)
```

```python
## encapsulation with getter and setter 

class Person:
    def __init__(self, name, age):
        self.__name = name # private access modifier or variable
        self.__age = age # private variable

## getter method for name
    def get_name(self):
        return self.__name

## setter method for name(it helps in changing the value)
    def set_name(self, name):
        self.__name = name

## getter method for age
    def get_age(self):
        return self.__age

## setter method for age(it helps in changing the value)
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age can't be negative")             



person=Person("John", 30)

## access nd modify private variables using getter and setter

print(person.get_name())
print(person.get_age())

person.set_name("Jane")
person.set_age(25)

print(person.get_name())
print(person.get_age())

person.set_age(-5)
```

## Abstraction

Abstraction is the concept of hiding the complex implementation etails and showing only the necessary features of an object. this helps in reducing programming programming complexity and effort.

example:Hair Dryer: You only use the on/off switch. The mechanics behind the fan and heating coil are hidden.

```python
from abc import ABC, abstractmethod

## abstract base class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod   #it gives necessary functionality with hiding complex implementation details
    def start_engine(self):
        pass      ## empty body

class Car(Vehicle): #inheritance
    def start_engine(self):
        print("The car engine is started")   

def operate_vehicle(vehicle): #polymorphism
    vehicle.start_engine()

car=Car() 
operate_vehicle(car)           

```
