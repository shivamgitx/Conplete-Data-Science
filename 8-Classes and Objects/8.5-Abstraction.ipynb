## Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object.This hels in reducing programming complexity and comfort.

```python
from abc import ABC,abstractmethod 

##Abstract class

class Vehicle(ABC):
    def drive(self): 
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self) :
        pass

class Car(Vehicle):
    def start_engine(self):
        print("The car engine is started")
         
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)    
```
