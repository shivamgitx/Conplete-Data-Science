## Polymorphism

Polymorphism is a core concept of OOP that allows objects of different classes to be treated as objects of a common superclass . it provides a way to perform  single action in different forms . Polymorphism is typically achieved through method overriding and interfaces.


In Python, polymorphism is an object-oriented programming concept that allows a single interface (like a function, method, or operator) to handle different types of data or objects

Derived from Greek, it literally means "many forms". 

### MEthod Overriding

Method overriding allows a child class to provide  specific implementation of a method that is already defind in its parent class.

```python
##Base Class
class Animal:
    def speak(self):
        return "Sound of th animal"
    
## Derived class 1
class Dog(Animal):
    def speak(self):
        return "Woof"    
    
## Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow"    

## Funtion that demonstrate polymorphism
def animal_speak(animal):
    print(animal.speak())


##object
dog=Dog()
cat=Cat()
print(dog.speak()) 
print(cat.speak())   
animal_speak(dog)
```

```python
## Polymorphism with function and method
## base class
class Shape():
    def area(self):
        return "The are of the shape"
    
## Derived class1 
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height
        
    def area(self):
        return self.width*self.height
    
## Derived class 2
#
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
        
    def area(self):
        return 3.14*self.radius*self.radius
    

## Function that demonstarte the polymorphism

def print_area(shape):
    print(shape.area())
    
## Objects
rectangle=Rectangle(10,20)
circle=Circle(5)

print_area(rectangle)
print_area(circle)
        


```

# Polymorphism with Abstract Base class

Abstract Base Classes(ABCs) are used to define common methods for a group 0f related objects . 
They can enforce hat derived class implement particular mrthods, promoting consistency accross different implementations.

```python
from abc import ABC, abstractmethod

## Define an abstract class

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

## Derived class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car Engine Started"

## Derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle Engine Started"


# function that demonstarte the polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())

## objects

car = Car()
motorcycle=Motorcycle()

start_vehicle(car)
```
