## Custom exception (Raise and Throw an exception) 

```python
class Error(Exception):
    pass

class dobException(Error):
    pass

```

```python
year=int(input("Enter your dob:"))
age=2024-year

try:
    if age <=30 and age>=20:
        print("the agr is valid so you can apply for the exam")
    else:
        raise dobException
    
except dobException:
    print ("Sorry your age is should be grater than 20 and less than 30")    
```
