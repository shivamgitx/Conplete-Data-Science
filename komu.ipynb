### Practical 1: Social Media Metrics Analysis

```python
# Import required library
import pandas as pd
```

```python
#Create dataset (as given in question)
data = {
    "Post_ID": ["P101", "P102", "P103", "P104", "P105"],
    "Date": ["01-03-2026", "02-03-2026", "03-03-2026", "04-03-2026", "05-03-2026"],
    "Likes": [120, 200, 150, 300, 180],
    "Comments": [30, 50, 25, 80, 40],
    "Shares": [20, 40, 30, 60, 35],
    "Reach": [1000, 1800, 1400, 2500, 1600],
    "Impressions": [1500, 2500, 2000, 3500, 2200]
}
```

```python
# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Calculate Engagement Rate
# Formula: (Likes + Comments + Shares) / Reach * 100
df["Engagement_Rate (%)"] = (
    (df["Likes"] + df["Comments"] + df["Shares"]) / df["Reach"]
) * 100
```

```python
# Display dataset with calculated metric
print("📊 Social Media Dataset with Engagement Rate:\n")
print(df)
```

```python
# Step 5: Generate summary report
print("\n📈 Summary Report:\n")

# Average Engagement Rate
avg_engagement = df["Engagement_Rate (%)"].mean()
print(f"Average Engagement Rate: {avg_engagement:.2f}%")

# Total Reach and Impressions
total_reach = df["Reach"].sum()
total_impressions = df["Impressions"].sum()

print(f"\nTotal Reach: {total_reach}")
print(f"Total Impressions: {total_impressions}")
```

### Practical 3: Identify Positive and Negative Words

```python
# Sample dataset given in question
data = [
    "I love this product, it's amazing!",
    "Worst service ever, very disappointed",
    "It is okay, not great not bad"
]
```

```python
# Define positive and negative words
positive_words = ["love", "amazing", "good", "excellent", "great"]
negative_words = ["bad", "worst", "disappointed"]
```

```python
# check
for text in data:
    print("\nText:", text)      # Print the original sentence
    
    for word in text.lower().split():   # Convert sentence to lowercase and split into words
        word = word.strip("!,.'")       # Remove punctuation like ! , . '
        
        if word in positive_words:     # Check if word is in positive words list
            print(word, "-> Positive")
        if word in negative_words:     # Check if word is in negative words list
            print(word, "-> Negative")
```

### Practical 4: Text Preprocessing in Python

```python
# Import libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# Sample dataset
data = [
    "I love this product, it's amazing!",
    "Worst service ever, very disappointed",
    "It is okay, not great not bad"
]
```

```python
# Initialize
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Process text
for text in data:
    
    tokens = word_tokenize(text.lower())  # Tokenization
    
    filtered = []
    for word in tokens:
        if word.isalpha() and word not in stop_words:  # Stopword removal
            filtered.append(word)
    
    lemmatized = []
    for word in filtered:
        lemmatized.append(lemmatizer.lemmatize(word))  # Lemmatization
    
    print("\nOriginal:", text)
    print("Cleaned:", " ".join(lemmatized))
```

### Practical 5: Sentiment Analysis using Python

```python
from textblob import TextBlob

# Sample dataset
data = [
    "I love this product, it's amazing!",
    "Worst service ever, very disappointed",
    "It is okay, not great not bad"
]
```

```python
# Process each text
print("Text\t\t\t\tSentiment\tScore")

for text in data:
    
    # Create TextBlob object
    blob = TextBlob(text)
    
    # Get polarity score (-1 to 1)
    score = blob.sentiment.polarity

    # Classify sentiment
    if score > 0:
        sentiment = "Positive"
    elif score < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    # Print result
    print(f"{text[:30]}...\t{sentiment}\t{score:.2f}")
```

### Practical 6: Twitter Sentiment Analysis (Mini Project)

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# Step 1: Sample Twitter Data

data = {
    "Tweet": [
        "I love this phone, amazing features!",
        "Worst experience ever, very bad service",
        "It is okay, nothing special",
        "Excellent product, very happy",
        "Not good, disappointed"
    ]
}

df = pd.DataFrame(data)
```

```python
# Step 2: Cleaning (basic)

df["Clean_Tweet"] = df["Tweet"].str.lower()
```

```python
# Step 3: Sentiment Analysis

def get_sentiment(text):
    score = TextBlob(text).sentiment.polarity
    
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Clean_Tweet"].apply(get_sentiment)
```

```python
# Step 4: Count sentiments

sentiment_count = df["Sentiment"].value_counts()
```

```python
# Step 5: Plot graph
plt.figure()
sentiment_count.plot(kind="bar")

plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

### Practical 7: Machine Learning vs Deep Learning Comparison

```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Step 1: Create dataset
data = {
    "Study_Hours": [2, 5, 1, 6, 3],
    "Attendance": [60, 75, 50, 85, 65],
    "Previous_Score": [50, 65, 40, 70, 55],
    "Final_Result": [0, 1, 0, 1, 0]
}
df = pd.DataFrame(data)
```

```python
# Step 2: Define X (input) and y (output)
X = df[["Study_Hours", "Attendance", "Previous_Score"]]
y = df["Final_Result"]
```

```python
# Step 3: Split data (train & test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Step 4: Train model (Logistic Regression)
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Step 5: Predict & check accuracy

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

# Output
print("Predicted:", y_pred)
print("Actual:", list(y_test))
print("Accuracy:", accuracy)
```

```python
# Step 4: Extract Headings
print("Headings:")

# Extract all headings (h1, h2)
for heading in soup.find_all(["h1", "h2"]):
    print(heading.text)
```
