## Importing Modules in Python :Modules and Packages

In Python, modules and packages help organise and reuse code. Here's a comprehensive guide on how to import them.

```python
import math
math.sqrt(4)
```

```python
from math import sqrt,pi
print(sqrt(4))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(5))
print(pi)
```

```python
from packages.maths import addition
addition(2,3)
```

```python
from packages import maths
addition(2,5)
```

```python
from packages import maths
maths.addition(2,5)
```
